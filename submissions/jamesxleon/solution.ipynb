```python
# Read the csv file and create pd dataframe
import pandas as pd 
df = pd.read_csv('spotify-2023.csv', encoding='latin1')
df.head()
```

```python
# Number of songs in the dataset (rows)
print("Number of songs in the dataset: ", df.shape[0])
```

```python
df_E_key = df[df['key'] == 'E']
df_E_key.head()
```

```python
print("Number of songs in the dataset with key E: ", df_E_key.shape[0])
```

```python
# Name of the columns in df
print("Name of the columns in df: ", df.columns)
```

```python
#Given a column name, find the value with the most occurrences
def most_occurrences(column_name, df=df):
    return df[column_name].value_counts().idxmax()

print("Total number of songs: ", df.shape[0])
print("Most common artist: ", most_occurrences("artist(s)_name"), " with ", df["artist(s)_name"].value_counts().max(), " occurrences")
print("Most common key: ", most_occurrences("key"), " with ", df["key"].value_counts().max(), " occurrences")
print("Most common mode: ", most_occurrences("mode"), " with ", df["mode"].value_counts().max(), " occurrences")
print("Most common bmp: ", most_occurrences("bpm"), " with ", df["bpm"].value_counts().max(), " occurrences")
```
